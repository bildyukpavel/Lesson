```python
Имя: В этом столбце указано имя пациента, связанного с медицинской картой.
Возраст: Возраст пациента на момент поступления, выраженный в годах.
Пол: Указывает пол пациента: «Мужской» или «Женский».
Группа крови: Группа крови пациента, которая может быть одной из распространенных (например, «A+», «O-» и т. д.).
Состояние здоровья: В этом столбце указывается основное заболевание или диагноз, связанный с пациентом, например, «диабет», «гипертония», «астма» и другие.
Дата поступления: Дата, когда пациент был госпитализирован в медицинское учреждение.
Врач: Имя врача, ответственного за лечение пациента во время его госпитализации.
Больница: Указывает медицинское учреждение или больницу, куда был госпитализирован пациент.
Страховая компания: В этом столбце указана страховая компания пациента, которая может быть одной из нескольких, включая «Aetna», «Blue Cross», «Cigna», «UnitedHealthcare» и «Medicare».
Сумма к оплате: сумма денег, выставленная пациенту за медицинские услуги во время его госпитализации. Выражается в виде числа с плавающей запятой.
Номер палаты: Номер палаты, в которой пациент проживал во время госпитализации.
Тип госпитализации: Указывает тип госпитализации, который может быть «экстренной», «плановой» или «срочной», в зависимости от обстоятельств госпитализации.
Дата выписки: Дата, когда пациент был выписан из медицинского учреждения, определяемая на основе даты поступления и случайного количества дней в пределах реалистичного диапазона.
Лекарственные препараты: Указывается лекарственный препарат, назначенный или введенный пациенту во время госпитализации. Примеры: «Аспирин», «Ибупрофен», «Пенициллин», «Парацетамол» и «Липитор».
Результаты анализов: Описывает результаты медицинского анализа, проведенного во время госпитализации пациента. Возможные значения: «Норма», «Отклонение от нормы» или «Неопределенный результат», указывающие на исход анализа.
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
X = dataset.drop(columns = ['Test Results'], axis = 1)
y = dataset['Test Results']
accuracy_scores_list = []
f1_scores_list = []
model_names_list = []
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
```

```python
#Logistics regression
```

```python
logistics = LogisticRegression(solver = 'liblinear')
logistics.fit(X_train,y_train)
logistics_predict = logistics.predict(X_test)
logistics_accuracy = accuracy_score(y_test, logistics_predict)
logistics_f1_score = f1_score(y_test, logistics_predict, average = None)
accuracy_scores_list.append(logistics_accuracy)
f1_scores_list.append(logistics_f1_score)
model_names_list.append('Логистическая регрессия')
```

```python
# knn
```

```python
knn = KNeighborsClassifier(n_neighbors= 14)
knn.fit(X_train, y_train)
knn_predict = knn.predict(X_test)
knn_accuracy = accuracy_score(y_test, knn_predict)
knn_f1_score = f1_score(y_test, knn_predict, average = None)
accuracy_scores_list.append(knn_accuracy)
f1_scores_list.append(knn_f1_score)
model_names_list.append('Метод К ближайщих')
```

```python
#Tree
```

```python
Tree = tree.DecisionTreeClassifier(max_depth=4,min_samples_leaf= 1, min_samples_split= 15)
Tree.fit(X_train, y_train)
Tree_predict = Tree.predict(X_test)
Tree_accuracy = accuracy_score(y_test, Tree_predict)
Tree_f1_score = f1_score(y_test, Tree_predict, average = None)
accuracy_scores_list.append(Tree_accuracy)
f1_scores_list.append(Tree_f1_score)
model_names_list.append('Дерево решений')
```

```python
plt.figure(figsize=(30,15))
tree.plot_tree(Tree) 
plt.show()
```

```python
accuracy_results = pd.DataFrame({
    'Model': model_names_list,
    'Accuracy': accuracy_scores_list,
    'Accuracy (%)': [f"{a * 100:.2f}%" for a in accuracy_scores_list]
})


accuracy_results = accuracy_results.sort_values(by='Accuracy', ascending=False)

print("\n--- Сводная Таблица Результатов ---")
print(accuracy_results.to_string(index=False))
```

```python
for model_name, f1_score in zip(model_names_list, f1_scores_list):
    print(model_name,':', f1_score)
```
