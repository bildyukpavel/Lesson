```python
#2

def check_pin(x:int):
    if len(str(x)) == 4 or len(str(x)) == 6:
        return True
    else:
        return False
    

check_pin(123456)
```

```python
#3

def check_number(x1:list):
    result = []
    for i in x1:
        if i ** 0.5  == int(i ** 0.5) :
            result.append(i)
    return result

check_number([25,30,100])
```

```python
#4

def unique_in_order(lst:list):
    new_lst = []
    for q in range(len(lst)):
        if (q == len(lst) - 1 or lst[q] != lst[q + 1] ) and (lst[q] != lst[q - 1] or q == 0 ): 
            new_lst.append(lst[q])
    return new_lst

unique_in_order([6,1,1,1,2,2,3,1,5,6])
```

```python
#5

def check_slov(str1:str):
    str2 = str1.split()
    v = str2[0]
    for i in str2:
        if len(i) < len(v):
            v = i
    return v
check_slov('avtomobil dom doroga derevo qw')
```

```python
#6

def rec(n):
    if n <= 1:return 1
    return  n * rec(n - 1)


def cifrovoi_koren(n):
    while n > 9:
        n = sum(int(kor) for kor in str(n))
    return n

cifrovoi_koren(rec(4))
```

```python
#7


def slovo(s:str):
    if len(s) % 2 == 0:
        y1 = len(s) // 2 
        y2 = s[y1 - 1],s[y1]
        print(f'Длинна слова четная,средние символы: {y2}')
    else:
        y1 = len(s) // 2 
        y2 = s[y1]
        print(f'Длинна слова не четная ,средний символ: {y2}')
        
slovo('копать')
```

```python
#8

def prostoe_n(n:int):
    if (2 == 2 and n % 2 == 0) and (n % 3 == 0 and 3 == 3):
        return False
    else:
        return True

prostoe_n(163)
```

```python
#9

def sum_celih_chisel(z:list) -> int:
    result_sum:int = 0
    for i in z:
        result_sum+= i
    if result_sum % 2 == 0:
        print(f'Результат: {result_sum}.Сумма чисел - четная.')
    else:
        print(f'Результат: {result_sum}.Сумма чисел - нечетная.')
        
sum_celih_chisel([1,2,3,12,100,-20])
```
