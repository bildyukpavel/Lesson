```python
X = dataset.drop(columns = ['price_range'])
y = dataset['price_range']
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42, test_size = 0.2)
```

```python
#logistic model
```

```python
logistics = LogisticRegression(solver='liblinear')
logistics.fit(X_train,y_train)
logistics_predict = logistics.predict(X_test)
```

```python
logistics_acc = accuracy_score(y_test,logistics_predict)
print("Logistic Accuracy: ",logistics_acc)
```

```python
logistics_precision = precision_score(y_test, logistics_predict, average = None)
print("Logistic Precision: ",logistics_precision)
```

```python
logistics_recall = recall_score(y_test, logistics_predict, average = None)
print("Logistic Recall: ",logistics_recall)
```

```python
logistics_f1_score = f1_score(y_test, logistics_predict, average = None)
print("Logistic f1: ",logistics_f1_score)
```

```python
svc_acc = accuracy_score(y_test,svc_predict)
print("SVC Accuracy: ",svc_acc)
```

```python
svc_precision = precision_score(y_test, svc_predict, average = None)
print("SVC Precision: ",logistics_precision)
```

```python
svc_recall = recall_score(y_test, svc_predict, average = None)
print("SVC Recall: ", svc_recall)
```

```python
svc_f1_score = f1_score(y_test, svc_predict, average = None)
print("SVC F1 score: ",svc_f1_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 13)
knn.fit(X_train,y_train)
knn_predict = knn.predict(X_test)
```

```python
knn_acc = accuracy_score(y_test,knn_predict)
print("KNN Accuracy: ", knn_acc)
```

```python
knn_precision = precision_score(y_test, knn_predict, average = None)
print("KNN Precision: ", knn_precision)
```

```python
knn_recall = recall_score(y_test, knn_predict, average = None)
print("KNN Recall: ", knn_recall)
```

```python
knn_f1 = f1_score(y_test, knn_predict, average = None)
print("KNN F1: ", knn_f1)
```

```python
import tensorflow as tf
```

```python
import keras
```

```python
from tensorflow.keras import layers
```

```python
from keras import Input
from keras.models import Sequential
from keras.layers import Flatten, Dense
```

```python
model = keras.Sequential([
    Input(shape=(20,)),
    Flatten(),
    Dense(128,activation = 'relu'),
    Dense(4)
])
```

```python
model.compile(optimizer = 'adam',
             loss = tf.keras.losses.MeanSquaredError(
                reduction="sum_over_batch_size"),
                metrics = ['mae']
             )
```

```python
model.fit(X, y, epochs=250, verbose=1)
```

```python
model.evaluate(X, y, verbose = 2)
```
