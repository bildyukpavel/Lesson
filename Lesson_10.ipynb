```python
# "Умный Дом: Сценарии" (Поведенческое наследование)
# Концепция: Различные сценарии (режимы) работы умного дома, которые по-разному реагируют на одни и те же события.
# Задача:
# Базовый класс: HomeScenario
# Методы:
# on_motion_detected(room) — выводит f"Движение в {room}. Запись началась."
# on_door_opened(door) — выводит f"{door} открыта. Оповещение отправлено."
# Классы-наследники:
# AwayScenario (Режим "Никого нет дома"):
# Переопределяет on_motion_detected — включает сирену и записывает видео.
# Переопределяет on_door_opened — блокирует дверь и звонит в полицию.
# SleepScenario (Режим "Сон"):
# Переопределяет on_motion_detected — если движение в коридоре, включает ночной свет на 10%.
# Переопределяет on_door_opened — если открыта входная дверь, включает сигнализацию.
# PartyScenario (Режим "Вечеринка"):
# Переопределяет on_motion_detected — игнорирует.
# Переопределяет on_door_opened — проигрывает звук "Гость пришел!".
```

```python
class HomeScenario:
    def __init__(self, motion: bool, door: bool, room: str):
        self.get__motion = motion
        self.get__door = door
        self.get__room = room
        self.get__dict_rooms = {"Кухня": 1, "Гараж": 2, "Туалет": 3, "Гостинная": 4,"Главный вход" : 5, "Коридор" : 6}

   
    def on_motion_detected(self):
        temp = self.get__dict_rooms.keys()
        
        for i in range(len(temp)):
            if self.get__room in temp and self.get__motion == False:
                print(f'Движение в {self.get__room}.Запись началась')
                
            break
        

    def on_door_opened(self):
        temp = self.get__dict_rooms.keys()
        
        for i in range(len(temp)):
            if self.get__room in temp and self.get__motion == False:
                print(f'Дверь {self.get__room} открыта.оповещение отправленно')
                
            break


class AwayScenario(HomeScenario):                                               #Режим : Никого нет дома
    
    def __init__(self, motion: bool, door: bool, room: str):
        super().__init__(motion, door, room)
        self.__temp = self.get__dict_rooms.keys()
        
    
    def on_motion_detected(self):
        for i in range(len(self.__temp)):
            if self.get__room in self.__temp and self.get__motion == False:
                print(f"Сирена включена. Запись началась")
                
            break

    def on_door_opened(self):
       for q in range(len(self.__temp)): 
           if self.get__room in self.__temp and self.get__door == False:
               self.get__door = True
               print(f"{self.get__room}: открыта Дверь.\nБлокировка двери.\nЗвонок в полицию.")
               



class SleepScenario(HomeScenario):                                               #Режим сна
    
    def __init__(self, motion: bool, door: bool, room: str):
        super().__init__(motion, door, room)

    def on_motion_detected(self):
        temp = self.get__dict_rooms.keys()
        room1 = "Коридор"
        
        if room1 == self.get__room and self.get__motion == False :
            print("Включен ночной свет на 10%")


    def on_door_opened(self):
        temp = self.get__dict_rooms.keys()
        room2 = "Главный вход"
        
        if room2 == self.get__room and self.get__door == False :
            print(f"{self.get__room}: открыта Дверь. Включена сигнализация.")
            
            



class PartyScenario(HomeScenario):                                              # Режим вечеринки
    def __init__(self, motion: bool, door: bool, room: str):
        super().__init__(motion, door, room)

    def on_door_opened(self):
        room2 = "Главный вход"
        
        if room2 == self.get__room and self.get__door == False :
            print(f"Пришел гость!")






    

#dict_rooms = {"Кухня": 1, "Гараж": 2, "Туалет": 3, "Гостинная": 4,"Главный вход" : 5, "Коридор" : 6}
# motion = True  движения нет,False - Есть движение 
#door = True ,дверь закрыта ,False - дверь открыта
```
