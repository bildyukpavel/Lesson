```python
class Shape:

    def __init__(self,name,atribute):
        self.__name = name
        self.__atribute = atribute
       


    
    def get_area(self):
        pass
        
    
    def get_perimetr(self):
        pass
   
    def get_name(self):
        print(f'Тип фигуры: {self.__name}')


    def display_info(self):
        print(f'Информация о фигуре :\nАтрибут фигуры: {self.__atribute} \nДанные фигуры:\nПлощадь - {self.get__area},\nПериметр - {self.get__perimetr}')
       

class Circle(Shape):

    def __init__(self,name,atribute,r):
        super().__init__(name,atribute)
        self.__name = name
        self.__atribute = atribute
        self.__r = r


    @property
    def get__area(self):
        result = (self.__r ** 2) * 3,14
        if self.__r > 0:
            return result
        else:
            return 'Отриательное значение.Введите положительное значение'

    @property
    def get__perimetr(self):
        result = 2 * (self.__r + self.__r)
        if self.__r > 0:
            return result
        else:
            return 'Отриательное значение.Введите положительное значение'
        



class Rectangle(Shape):

    def __init__(self,name,atribute,l,b):
        super().__init__(name,atribute)
        self.__name = name
        self.__atribute = atribute
        self.__lenght = l
        self.__width = b

    @property
    def get__area(self):
        if self.__lenght  > 0 < self.__width :
            result = self.__lenght * self.__width
            return result
        else:
            return 'Отриательное значение.Введите положительное значение'
    
    @property
    def get__perimetr(self):
        if self.__lenght > 0 < self.__width :
            result = 2 * (self.__lenght + self.__width)
            return result
        else:
            return 'Отриательное значение.Введите положительное значение'



class Triangle(Shape):

    def __init__(self,name,atribute,side_a,side_b,side_c):
        super().__init__(name,atribute)
        self.__name = name
        self.__atribute = atribute
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
        self.__area_p = (self.__side_a + self.__side_b + self.__side_c ) / 2

    @property
    def get__area(self):
        if self.__side_a > 0 and self.__side_c > 0 and self.__side_c > 0:
            if self.__side_a + self.__side_b > self.__side_c  and self.__side_b + self.__side_c > self.__side_a:
                result = (self.__area_p * (self.__area_p - self.__side_a) * (self.__area_p - self.__side_b) * (self.__area_p * self.__side_c) ** 0.5)
                return result 
            else:
                return 'Сумма двух сторон меньше третей.Введите коректные стороны'
            
        else:
            return 'Отриательное значение.Введите положительное значение'

    @property
    def get__perimetr(self):
        if self.__side_a > 0 and self.__side_c > 0 and self.__side_c > 0:
            if self.__side_a + self.__side_b > self.__side_c  and self.__side_b + self.__side_c > self.__side_a:
                result = result = self.__side_a + self.__side_b + self.__side_c
                return result 
            else:
                return 'Сумма двух сторон меньше третей.Введите коректные стороны'
            
        else:
            return 'Отриательное значение.Введите положительное значение'





class Square(Rectangle):

    def __init__(self,name,atribute,lenght):
        super().__init__(name,atribute,lenght,lenght)





circle = Circle(name = "Круг",atribute = "Радиус ",r = 3)
rectangle = Rectangle(name = "Прямоугольник",atribute = "Длинна,Ширина ",l = 3,b = 5)
triangle = Triangle(name = "Триугольник",atribute = "Стороны а,b,c ",side_a = 3,side_b = 4,side_c = 5)
square = Square(name = 'Квадрат',atribute = 'Длинна',lenght = 4)
```

```python
class Employee:

    def __init__(self,id,name,job_title,data_of_employment):
        self.__id = id
        self.__name = name
        self.__job_title = job_title
        self.__data_of_employment = data_of_employment
    
        

    def salary_calculate(self):
        calculate = self.get_salary_calculate * 0.13
        print(f'Зарплата составляет: {self.get_salary_calculate - calculate}')

    def vacation_pay_calculate(self):
        calculate = self.get_vacation_pay_calculate * 0.13
        print(f'Отпускные составляют: {self.get_vacation_pay_calculate - calculate}')

    def nalog(self):
        result = self.get_vacation_pay_calculate * 0.13
        print(f'Налог составялет: {result}')

    
    def info_employee(self):
        print(f'id сотрудника: {self.__id}. Имя сотрудника :{self.__name}. Должность: {self.__job_title}. Дата приема: {self.__data_of_employment}')



class FullTimeEmployee(Employee):

    def __init__(self,id,name,job_title,data_of_employment):
        super().__init__(id,name,job_title,data_of_employment)
        self.__id = id
        self.get__name = name
        self.__job_title = job_title
        self.__data_of_employment = data_of_employment
        self.__fix_salary = 9887
        self.__bonuse = 0.30
        self.__health_insurance = True

    @property
    def get_salary_calculate(self):
        return self.__fix_salary

    @property
    def get_vacation_pay_calculate(self):
        calculate_bonus = self.__fix_salary * self.__bonuse
        result = ((calculate_bonus + self.__fix_salary) + self.__fix_salary + self.__fix_salary) // 3
        return result







class PartTimeEmployee(Employee):

    def __init__(self,id,name,job_title,data_of_employment,hourse_worked):
        super().__init__(id,name,job_title,data_of_employment)
        self.__id = id
        self.__name = name
        self.__job_title = job_title
        self.__data_of_employment = data_of_employment
        self.__salary_of_hourse = 89
        self.__hourse_worked = hourse_worked
        self.__health_insurance = None
        self.__salary = self.__hourse_worked * self.__salary_of_hourse



    @property
    def get_salary_calculate(self):
        return self.__salary

    @property
    def get_vacation_pay_calculate(self):
        result =  (self.__salary + self.__salary + self.__salary) / 3
        return result

class CommissionEmployee(Employee):

    def __init__(self,id,name,job_title,data_of_employment,sales_volume):
        super().__init__(id,name,job_title,data_of_employment)
        self.__id = id
        self.__name = name
        self.__job_title = job_title
        self.__data_of_employment = data_of_employment
        self.__fixed_solary = 5432
        self.__sales_volume = sales_volume
        self.__fixed_sales_volume = 100
                
    @property
    def get_salary_calculate(self):
        if self.__sales_volume > self.__fixed_sales_volume:
            calculate_procent = self.__fixed_solary * 0.20
            return self.__fixed_solary + calculate_procent
        else:
            return self.__fixed_solary


    @property
    def get_vacation_pay_calculate(self):
        result =  (self.get_salary_calculate + self.get_salary_calculate + self.get_salary_calculate) / 3
        return result


    


class Manager(FullTimeEmployee):

    def __init__(self,id,name,job_title,data_of_employment):
        super().__init__(id,name,job_title,data_of_employment)
        self.__money_for_team_training = 12345
        self.__bonus_managment = 0.30
        self.__fix_salary = 9887


    @property
    def get_salary_calculate(self):
        result =  self.__fix_salary * self.__bonus_managment 
        return self.__fix_salary + result

    @property
    def get_vacation_pay_calculate(self):
        result =  (self.get_salary_calculate + self.get_salary_calculate + self.get_salary_calculate) / 3
        return result




    
employee_ft = FullTimeEmployee(id = 1,name = 'Андрей',job_title = "Штатный сотрудник", data_of_employment = '21,12,2012')
employee_pt = PartTimeEmployee(id = 2,name = 'Виктор',job_title = "Сотрудник на неполный день", data_of_employment = '11,10,2023',hourse_worked = 102)
employee_ce = CommissionEmployee(id = 3,name = 'Евгений',job_title = "комиссионный сотрудник", data_of_employment = '14,02,2019',sales_volume = 102)
employee_m = Manager(id = 4,name = 'Кирилл',job_title = "Менеджер", data_of_employment = '14,03,2012')
```
