```python
summarizer_list = ['bart-large-cnn', 'distilbart-xsum-1-1', 'MEETING_SUMMARY']
text_list = ['sport_news', 'sport_news_2', 'buisness_news']

rouge_scor = [
    rouge_scores_1_1['rouge1'], rouge_scores_1_1['rouge2'],
    rouge_scores_1_2['rouge1'], rouge_scores_1_2['rouge2'],
    rouge_scores_1_3['rouge1'], rouge_scores_1_3['rouge2'],
    rouge_scores_2_1['rouge1'], rouge_scores_2_1['rouge2'],
    rouge_scores_2_2['rouge1'], rouge_scores_2_2['rouge2'],
    rouge_scores_2_3['rouge1'], rouge_scores_2_3['rouge2'],
    rouge_scores_3_1['rouge1'], rouge_scores_3_1['rouge2'],
    rouge_scores_3_2['rouge1'], rouge_scores_3_2['rouge2'],
    rouge_scores_3_3['rouge1'], rouge_scores_3_3['rouge2']
    
             ] 
bleu_scor = [
    bleu_scores_1_1['bleu'], bleu_scores_1_1['precisions'],
    bleu_scores_1_2['bleu'], bleu_scores_1_2['precisions'],
    bleu_scores_1_3['bleu'], bleu_scores_1_3['precisions'],
    bleu_scores_2_1['bleu'], bleu_scores_2_1['precisions'],
    bleu_scores_2_2['bleu'], bleu_scores_2_2['precisions'],
    bleu_scores_2_3['bleu'], bleu_scores_2_3['precisions'],
    bleu_scores_3_1['bleu'], bleu_scores_3_1['precisions'],
    bleu_scores_3_2['bleu'], bleu_scores_3_2['precisions'],
    bleu_scores_3_3['bleu'], bleu_scores_3_3['precisions']
]
```

```python
idx = 0
for summar in summarizer_list:
    for text in text_list:
        rouge1 = rouge_scor[idx]     
        rouge2 = rouge_scor[idx + 1] 
        bleu = bleu_scor[idx]
        bleu_precision = bleu_scor[idx + 1]
        print(f'Модель: {summar}, Текст : {text}. rouge: {rouge1:.2f}, rouge2: {rouge2:.2f}')
        print(f'Модель: {summar}, Текст : {text}.  bleu: {bleu:.2f}, bleu_precision: {[f'{prec:.2f}' for prec in bleu_precision]}')
        idx += 2
```
