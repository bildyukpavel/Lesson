```python
import sqlalchemy
```

```python
import sqlalchemy as db
```

```python
engine = db.create_engine('sqlite:///mylessondb.db')
```

```python
conn = engine.connect()
```

```python
metadata = db.MetaData()
```

```python
dir(db)
```

```python
persons = db.Table(
    'person',metadata,
    db.Column('person_id', db.Integer,primary_key = True),
    db.Column('name', db.Text),
    extend_existing=True
)
persons
```

```python
metadata.create_all(engine)
```

```python
cinemas = db.Table(
    'cinema', metadata,
    db.Column('id', db.Integer, primary_key = True),
    db.Column('person',db.Integer),
    db.Column('room', db.Integer),
    extend_existing=True
)
```

```python
metadata.create_all(engine)
```

```python
insertion1 = persons.insert().values([
    {'name' : 'Alex'},
    {'name' : 'John'}
])
```

```python
insertion2 = cinemas.insert().values([
    {'person' : 0, 'room' : 1},
    {'person' : 1, 'room' : 2}
])
```

```python
conn.execute(insertion1)
```

```python
conn.execute(insertion2)
```

```python
update_query = db.update(cinemas).where(cinemas.columns.person == 2).values(room = 5)
```

```python
conn.execute(update_query)
```

```python
select_all_quere_2 = db.select(cinemas)
select_result_2 = conn.execute(select_all_quere_2)
select_result_2.fetchall()
```

```python
order1 = db.select(persons).order_by(persons.columns.name)
select_result = conn.execute(order1)
select_result.fetchall()
```

```python
join = db.select(persons,cinemas).join(cinemas,cinemas.columns.person == persons.columns.person_id)
```

```python
conn.execute(join)
```
